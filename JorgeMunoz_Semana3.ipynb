Instrucciones:

En una empresa, se necesita gestionar la información de los empleados, incluyendo su nombre, departamento y salario. Crea un programa en Python que permita realizar las siguientes acciones:

Agregar empleados: El programa debe permitir al usuario ingresar el nombre, departamento y salario de un empleado, y luego agregar esta información a un diccionario que almacene todos los empleados.
Mostrar información de un empleado: El programa debe permitir al usuario ingresar el nombre de un empleado y mostrar en pantalla su departamento y salario.
Actualizar información de un empleado: El programa debe permitir al usuario ingresar el nombre de un empleado y actualizar su departamento y salario.
Eliminar un empleado: El programa debe permitir al usuario ingresar el nombre de un empleado y eliminar su información del diccionario.
Mostrar base de datos: Se muestra de manera ordenada y estructurada el contenido de toda la base de datos.
Salir del programa: esta opción da por terminada la ejecución del programa y permite salir del mismo.

## Imports

```python
from tabulate import tabulate
import pandas as pd
```

## Declare

```python
dict_empleados=dict()
lst_empleado=[]
```

## Init

```python
dict_empleados={
    "Jorge":["RH",7500],
    "Moises":["Gerencia",17500],
    "Rene":["Operaciones",9000]
}
```

## Modulos

### Modulo AgregarEmpleado

```python
def fn_Agregar(dic,nombre,dpto,salario):
    try:
        if dic.get(nombre) is None:
            dic[nombre]=[dpto,salario]
            print(f"El empleado {nombre} fue agregado exitosamente")
            print("\n")
        else:
            print(f"Ya existe un registro para {nombre}")
        return True
    except Exception as val:
        print(f"Ocurrio un error: ---- {val}")
        return False
```

### Modulo ImprimirEmpleado

```python
def fn_Imprimir(dic,nombre):
    try:
        lst_empleado=[]
        if dic.get(nombre) is None:
            print(f"El empleado {nombre} no existe")
            print("\n")
        else:
            lst_empleado.append(dic.get(nombre))
            print(f"Datos del empleado: {nombre}")
            print("\n")
            print(tabulate(lst_empleado, headers=["Departamento","Salario"]))
        return True
    except Exception as val:
        print(f"Ocurrio un error: ---- {val}")
        return False
```

### Modulo ActualizarEmpleado

```python
def fn_Actualizar(dic,nombre,departamento,salario):
    try:
        lst_empleado=[]
        if dic.get(nombre) is None:
            print(f"El empleado {nombre} no existe")
            print("\n")
        else:
            dic.update({nombre:[departamento,salario]})
            lst_empleado.append([departamento,salario])
            print(f"Datos del empleado: {nombre} fueron actualizados")
            print("\n")
            print(tabulate(lst_empleado, headers=["Departamento","Salario"]))
        return True
    except Exception as val:
        print(f"Ocurrio un error: ---- {val}")
        return False
```

### Modulo EliminarEmpleado

```python
def fn_Eliminar(dic,nombre):
    try:
        ans=str(input("Seguro que desea eliminar los datos de {nombre} S/N:"))
        ans=ans.lower()
        if ans=="s":
            if dic.get(nombre) is None:
                print(f"El empleado {nombre} no existe")
                print("\n")
            else:
                dic.pop(nombre)
                print(f"Datos del empleado: {nombre} fueron eliminados")
                print("\n")
        else:
            print("Se cancelo la operacion de borrado")
        return True
    except Exception as val:
        print(f"Ocurrio un error: ---- {val}")
        return False
```

### Modulo ImprimirListadoEmpleados

```python
def fn_ImprimirTodo(dic):
    lst_empleados=[]
    try:
        for nombre,datos in dic.items():
            salario ="$"+str(datos[1])
            dpto=datos[0]
            lst_empleados.append([nombre,dpto,salario])        
        print(tabulate(lst_empleados, headers=["Nombre","Departamento","Salario"]))
        return True
    except Exception as val:
        print(f"Ocurrio un error: ---- {val}")
        return False
```

### Modulo Menu

```python
def menu():
    nombre=""
    
    departamento=""
    salario=0
    try:
        opcion=True
        while opcion:
            print ("""
            1.Agregar empleado
            2.Imprimir datos de empleado
            3.Actualizar datos de empleado
            4.Eliminar datos de empleado
            5.Imprimir todos los registros
            6.Salir
            """)
            opcion=int(input("Seleccione una opcion del menu "))
            
            if opcion==1: 
                print("\n Agregar empleado") 
                
                #####Input de variables####
                try:
                    nombre=str(input("Digite el nombre de el nuevo empleado "))
                    departamento=input("Digite el departamento ")
                    salario=float(input("Digite el salario "))
                except ValueError as val:
                    print(f"Revise el ingreso de datos {val}")
                    menu()
                #####Input de variables####
                
               #llamado de la funcion
                out=fn_Agregar(dict_empleados,nombre,departamento,salario)
                
                if out==True:
                    print("\n Registro agregado exitosamente")
                else:
                    print("Error durante el proceso")
                opcion=False
                menu()
                
            elif opcion==2:
                print("\n Imprimir datos de empleado")
                
                  #####Input de variables####
                try:
                    nombre=str(input("\n Digite el nombre de el empleado "))
                except ValueError as val:
                    print(f"Revise el ingreso de datos {val}")
                    menu()
                #####Input de variables####
                
                #llamado de la funcion
                out=(fn_Imprimir(dict_empleados,nombre))
                
                if out==False:
                    print("\n Ocurrio un error durante la ejecución del programa")
                
                opcion=False
                menu()
            elif opcion==3:
                print("\n Actualizar datos de empleado") 
                 #####Input de variables####
                try:
                    nombre=str(input("Digite el nombre de el empleado a actualizar "))
                    departamento=input("Digite el departamento ")
                    salario=float(input("Digite el salario "))
                except ValueError as val:
                    print(f"Revise el ingreso de datos {val}")
                    menu()
                #####Input de variables####
                
                #llamado de la funcion
                out=fn_Actualizar(dict_empleados,nombre,departamento,salario)
                
                if out==False:
                    print("\n Ocurrio un error durante la ejecución del programa")
                
                opcion=False
                menu()
            elif opcion==4:
                print("\n Eliminar datos de empleado") 
                
                #####Obtener codigo busqueda####
                try:
                    nombre=str(input("Digite el nombre de empleado a eliminar "))
                except ValueError as val:
                    print(f"Revise el ingreso de datos {val}")
                    menu()
                #############################################
                
                #llamado de la funcion
                out=fn_Eliminar(dict_empleados,nombre)
                if out==False:
                    print("\n Ocurrio un error durante la ejecución del programa")
                
                opcion=False
                menu()
            elif opcion==5:
                 #llamado de la funcion
                out=(fn_ImprimirTodo(dict_empleados))
                
                if out==False:
                    print("\n Ocurrio un error durante la ejecución del programa")
                opcion=False
                menu()
            elif opcion==6:
                print("\n Gracias por utilizar el sistema Empleados") 
                opcion=False
            elif opcion not in [1,2,3,4,5]:
                print("\n Seleccione una opcion valida entre 1 y 5") 
    except Exception as err:
        print(f"Unexpected {err}, {type(err)}")
        raise
    return True
```
